# Gun Deaths in U.S. Data

## Introducing the data

```python
import csv
f = open("guns.csv","r")
data = list(csv.reader(f))
print(data[:5])
```

## Removing Headers for a List of Lists

```python
headers = data[0]
data = data[1:]
print(headers)
print(data[:5])
```

## Counting Gun Deaths by Year

```python
years = [row[1] for row in data]
print(years)
```

```python
year_counts = {}
for year in years:
    if year not in year_counts:
        year_counts[year] = 0
    else:
        year_counts[year] += 1
print(year_counts)
```

## Exploring Gun Deaths by Month and Year

```python
import datetime
dates = [datetime.datetime(year=int(row[1]), month=int(row[2]), day=1) for row in data[1:]]

date_counts = {}
for date in dates:
    if date not in date_counts:
        date_counts[date] = 0
    else:
        date_counts[date] += 1
print(date_counts)
```

## Exploring Gun Deaths by Race and Sex

```python
sexes = [row[5] for row in data]
sex_counts = {}
for sex in sexes:
    if sex not in sex_counts:
        sex_counts[sex] = 0
    else:
        sex_counts[sex] += 1
        
races = [row[7] for row in data]
race_counts = {}
for race in races:
    if race not in race_counts:
        race_counts[race] = 0
    else:
        race_counts[race] += 1

print(sex_counts)
print(race_counts)
```

## Findings so far

It seems as though gun deaths affect females more than males in the US. They also seem to disproportionately affect minoriies, but it would be helpful to know the proportion of each race in the country.

Thee appears to be a seasonal correlation, with gun deaths peaking during the summer months and declining in the winter.
